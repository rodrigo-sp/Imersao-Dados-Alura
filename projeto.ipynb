```python
import numpy as np
# Compute the correlation matrix
corr = dados.loc[:,'g-0': 'g-50'].corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
corr_celular = dados.loc[:,'c-0': 'c-50'].corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
dados_resultados["n_moas"] = dados_resultados.drop("id", axis=1).sum(axis=1)
dados_resultados["ativo_moas"] = (dados_resultados["n_moas"]!=0)
display(dados_resultados)
```

```python
dados_combinados = pd.merge(dados, dados_resultados[["id", "n_moas", "ativo_moas"]], on='id')
display(dados_combinados)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x = dados_combinados.select_dtypes("float64")
y = dados_combinados["ativo_moas"]
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=376, stratify = y)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino,y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier("most_frequent")
# treinando o modelo
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
from sklearn.tree import DecisionTreeClassifier
x = dados_combinados.select_dtypes("float64")
y = dados_combinados["ativo_moas"]
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=376, stratify = y)

modelo_arvore = DecisionTreeClassifier(max_depth=3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(18,8), facecolor='k')
tree.plot_tree(modelo_arvore, ax=ax, fontsize=11, rounded=True, filled=True, feature_names = x_treino.columns,
              class_names = ["Não Ativado", "Ativado"])
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
x = dados_combinados.select_dtypes("float64")
y = dados_combinados["ativo_moas"]
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=376, stratify = y)

treino = []
teste = []
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth=i)
    modelo_arvore.fit(x_treino, y_treino)
    modelo_arvore.score(x_teste, y_teste)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
print(teste)
```

```python
print(treino)
```

```python
# plotando um gráfico das acurácias
plt.figure(figsize=(8,6))
x = np.arange(1,15) #range
plt.xlabel("Acurácia", fontsize=20)
plt.ylabel("Profundidade da árvore", fontsize=20)
plt.title("Acurácia dos dados teste e treino", fontsize=25, fontweight="bold")
plt.plot(x,teste, label="Teste", color="purple")
plt.plot(x,treino, label="Treino", color="black")
plt.tick_params(labelsize=20)
plt.legend(fontsize=20)
plt.show()
```
