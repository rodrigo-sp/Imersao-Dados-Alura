# Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

Dependendo o tipo de pesquisa é possível usar o mesmo controle para mais de um caso. Repare que o grupo de controle é um grupo onde não estamos aplicando o efeito de uma determinada droga. Então, esse mesmo grupo pode ser utilizado como controle para cada uma das drogas estudadas.

Um ponto relevante da base de dados que estamos trabalhando é que todos os dados de controle estão relacionados ao estudo de apenas uma droga.

```python
print(f"Total de dados {len(dados['id'])}\n")
print(f"Quantidade de drogas {len(dados.groupby(['droga', 'tratamento']).count()['id'])}\n")
```

```python
drogra_controle = list(filtro_com_controle["droga"].unique())
print(f"Todos os dados de controle estão relacionados com a droga: {drogra_controle[0]}")
```

Temos uma certa quantidade de culturas de células, no qual vamos análisar as 'respostas' dessas células para um determinado tipo de tratamento. Nestes tratamentos temos duas opções. Droga: colocando um determinado tipo de medicamento para cada cultura de célula. Controle: colocando as células em determinadas condições, por exemplo, ambientes diferentes "calor" e "frio" e ver as respostas a um deterinado intervalo de tempo. Aqui, "calor" e "frio" não necessariamente tem a ver com a nossa base de dados.

# Desafio 03: Proporção da classe tratamento 

```python
# vizualizando nossa série "tratamento"
# display(dados["tratamento"])
# display(dados["tratamento"].unique())
display(dados["tratamento"].value_counts(normalize="True"))
```

Portanto, para o nosso banco de dados, aproximadamente 93% dos tratamentos foram com drogas e que 7% foram com controle. A seguir visualizamos de forma gráfica.

```python
label_tratamento = dados["tratamento"].unique()
fig, ax = plt.subplots() # area de plotagem
destaca = (0, 0.1)
ax.pie(dados['tratamento'].value_counts(), explode = destaca, labels = label_tratamento, autopct="%1.1f%%", shadow=True, startangle=50)
ax.axis("equal")
plt.show()
```

# Análise com gráficos boxplot

```python
import seaborn as sns # importando a biblioteca

# características gerais da expressão gênica g-0
plt.figure(figsize=(8,6))
sns.boxplot(x='g-0', data=dados)
plt.tick_params(labelsize=18)
plt.show()
```

O gráfico acima apresenta de forma geral (todos características) da expressão gênica g-0. Podemos inferir que a mediana, isto é, a linha divisória que separa o conjunto de dados g-0 ao meio está no ponto zero. O primerio quartil (representa 25% da distribuição) está aproximadamente no ponto -0.5 e o terceiro quartil (representa 75% da distribuição) está aproximadamente no ponto 0.5. O limite inferior da expressão gênica g-0 está no ponto de -2, enquanto que o limite superior está no ponto 2. Fora desses dois extremos, temos os outliers, ou seja, são pontos “isolados” na nossa distribuição, sendo mais expressivos para um intervalo maior que o ponto 2 (limite superior) como é possível visualizar a concentração dos pontos.

```python
# comparando as expressões gênicas com droga e com controle para a expressão gênica g-0
plt.figure(figsize=(8,6))
sns.boxplot(y='g-0', x='tratamento', data=dados)
plt.tick_params(labelsize=15)
plt.show()
```

Do gráfico acima percebe-se que a mediana dos dados que representam a expressão gênica g-0 com droga e com controle são iguais a zero e seus limites (tanto superiores quanto inferiores) são muito próximos. A diferença está nos outliers, pois o tratamento com droga possui um maior número de pontos “isolados”. 

```python
# comparando as expressões gênicas g-0 para diferentes intervalos de tempo
plt.figure(figsize=(8,6))
sns.boxplot(y='g-0', x='tempo', data=dados)
plt.tick_params(labelsize=15)
plt.show()
```

Do gráfico acima percebe-se que as medianas para o gene g-0 com relação ao tempo em que a cultura dos mesmos foi submetida está no ponto zero. O primeiro ($\approx -0.8$) e o terceiro quartil ($\approx +0.8$) são muito próximos para todos os intervalos, e os limites superiores ($\approx 2$) e inferiores ($\approx -2$) são muitos próximos também. E por fim, a maioria dos outliers estão na região superior ao limite superior, tendo também outliers na região inferior ao limite inferior, mas em menor quantidade quando comparado aos dados "isolados" da outra região.

```python
# comparando a expressão gênica g-0 com relação as doses D1 e D2.
plt.figure(figsize=(8,6))
sns.boxplot(y='g-0', x='dose', data=dados)
plt.tick_params(labelsize=20)
plt.show()
```

Do boxplot acima mostra que temos uma pequena diferença nos limites superiores e inferiores para a mesma espressão gênica g-0 com relação as doses D1 e D2. Sendo D1 possuindo um limite superior e inferior maior em módulo (tomando só os valores absolutos) em relação a D2. As outras características: mediana, limite inferior, limite superior, outliers são bem parecidos.

```python
# comparando a expressão gênica g-0 com relação as doses D1 e D2.
for i in range (1,20):
    i = str(i)
    plt.figure(figsize=(8,6))
    plt.tick_params(labelsize=20)
    sns.boxplot(y=f'c-{i}', x='dose', data=dados)
    plt.show()
```

De forma geral, como é possível observar nos boxplot acima, as células possui medianas em zero; os limites superiores e inferiores são aproximadamente no ponto 2 e -2 respectivamente; e os outliers em sua maioria então no intervalo de -2 até aproximadamente -10.

```python
# cruzando as informações dose e tempo
pd.crosstab(dados["dose"], dados["tempo"])
```

Para visualizarmos melhor a tabela acima que já está muito clara, podemos fazer um gráfico.

```python
# armazenando a tabela em uma variável e fazendo a transposta
dados_dose_tempo = pd.crosstab(dados["dose"], dados["tempo"]).T
print(dados_dose_tempo)
```

```python
# filtrando os dados em relação a dose D1
dados_d1_tempo = dados_dose_tempo["D1"]
print(dados_d1_tempo)
```

```python
# filtrando os dados em relação a dose D2
dados_d2_tempo = dados_dose_tempo["D2"]
print(dados_d2_tempo)
```

```python
# gráfico comparativo da tabela dose e tempo
plt.figure(figsize=(8,6))
plt.tick_params(labelsize=20)
plt.title("Tipos de doses em relação aos intervalos de tempo", fontsize=25, fontweight="bold")
plt.xlabel("Dias", fontsize=20)
plt.ylabel("Frequências", fontsize=20)
tempo = ["24", "48", "72"]
plt.bar(tempo,dados_d1_tempo, color='yellow', alpha=.4, label="D1")
plt.bar(tempo,dados_d2_tempo, color='green', alpha=.4, label = "D2")
plt.legend(fontsize=10)
plt.show()
```

As variáveis dessa coluna são os mecanismos de ação da droga (composto). Assim, esse mecanismos é a interação bioquímica do composto com o alvo (pode ser a proteína de uma célula ou até mesmo de um patógeno que pode estar calsando a doença). Como é possível observar, os mecanismos de ação nos revela duas informalções, sendo: o alvo e a ação que ele tem sobre o alvo. Exemplificando, na primeira coluna temos "5-alpha_reductase_inhibitor"; 5-alpha_reductase é um enzima, portanto, o alvo e "inhibitor" é a ação, ou seja, esse mecanismo de ação está inibindo a enzima "5-alpha_reductase". De maneira geral "agonist" ativa o receptor e o "antagonist" bloqueia o receptor.

```python
# Verificando a coluna "acat_inhibitor"
display(dados_resultados["acat_inhibitor"].unique())
```

### Essa árvore de decisão apresenta inicialmente o nó raiz (gene "g-385") que é uma variável, a partir desta divide a  árvore em duas com  o objetivo de deixar o mais homogêno possível, por exemplo, de um lado uma maior probabilidade de experimentos que não foram ativos e de outro lado experimentos que foram ativados.

### Obs: 
    - gini: representa o quão homogêneo é esse conjunto de dados.
    - sample: o número de amostra.
    - class: é a decisão da nossa amostras "Ativado" ou "Não Ativado".

```python
from sklearn.tree import DecisionTreeClassifier
x = dados_combinados.select_dtypes("float64")
y = dados_combinados["ativo_moas"]
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=376, stratify = y)

treino = []
teste = []
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth=i)
    modelo_arvore.fit(x_treino, y_treino)
    modelo_arvore.score(x_teste, y_teste)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
print(teste)
```

```python
print(treino)
```

```python
# plotando um gráfico das acurácias
plt.figure(figsize=(8,6))
x = np.arange(1,15) #range
plt.xlabel("Acurácia", fontsize=20)
plt.ylabel("Profundidade da árvore", fontsize=20)
plt.title("Acurácia dos dados teste e treino", fontsize=25, fontweight="bold")
plt.plot(x,teste, label="Teste", color="purple")
plt.plot(x,treino, label="Treino", color="black")
plt.tick_params(labelsize=20)
plt.legend(fontsize=20)
plt.show()
```

Do gráfico acima é possível obeservar que para "baixas profundidades" na árvore de decisão, as acurácias teste e treino possuem valores parecidos (acurácia < 2.5) assim, a medida em que aumentamos a profundidade (nosso "$\textit{range}$") até um determinado ponto ($\approx 2.8$) as acurácias crescem, entretanto, a partir desse ponto a acurácia teste crece enquanto que a treino diminui. Então, a medida em que aumentamos a profundidade da árvore, o nosso modelo está pegando mais as características dos nossos dados de treino, assim, promovendo uma grande perda de qualidade em generalizar para dados não testados em nosso modelo. Esse processo em que nosso modelo fica muito bom em prever os dados de treino e perde a capacidade de generalização nos dados de teste é conhecido como "$\textit{overfitting}$ ". Uma forma de evitar o overfitting é criar várias árvores de decisão.
